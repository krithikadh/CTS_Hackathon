```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('data/hospital_readmissions.csv')
df.head()
```

```python
df.info()
```

```python
df.rename(columns={'time_in_hospital':'no_of_days',
                   'n_outpatient':'outpatient_visits',
                   'n_inpatient':'inpatient_visits',
                   'n_emergency':'emergency_visits',
                   'medical_specialty':'speciality',
                   'change':'diabetesmed_change'},
                   inplace=True)
df.head(10)
```

```python
print(df['age'].unique())
```

```python
df['age_grp'] = df['age'].replace({'[40-50)':4,
                                   '[50-60)':5,
                                   '[60-70)':6,
                                   '[70-80)':7,
                                   '[80-90)':8,
                                   '[90-100)':9})
df['age_grp'] = df['age_grp'].astype('category')
print('dtype :',df['age_grp'].dtype)
df[['age','age_grp']]
```

```python
df.drop('age', axis=1, inplace=True)
df.head()
```
