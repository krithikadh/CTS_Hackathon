```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('data/hospital_readmissions.csv')
df.head()
```

```python
df.info()
```

```python
df.rename(columns={'time_in_hospital':'no_of_days',
                   'n_outpatient':'outpatient_visits',
                   'n_inpatient':'inpatient_visits',
                   'n_emergency':'emergency_visits',
                   'medical_specialty':'speciality',
                   'change':'diabetesmed_change'},
                   inplace=True)
df.head(10)
```

```python
print(df['age'].unique())
df['age_grp'] = df['age'].replace({'[40-50)':'middle_aged',
                                   '[50-60)':'middle_aged',
                                   '[60-70)':'senior',
                                   '[70-80)':'senior',
                                   '[80-90)':'super_senior',
                                   '[90-100)':'super_senior'})
df['age_grp'] = df['age_grp'].astype('category')
print('dtype :',df['age_grp'].dtype)
df[['age','age_grp']]
```
